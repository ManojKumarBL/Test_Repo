```python
'''1. Analyzing the distribution of views by category:
          python
'''
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the dataset
df = pd.read_csv('data/youtube.csv')

df["views"] = pd.to_numeric(df['views'], errors='coerce')
df["category_id"] = pd.to_numeric(df['category_id'], errors='coerce')
plt.figure(figsize=(4, 4))
sns.barplot(data=df, x='category_id', y='views')
plt.title('Distribution of Views by Category')
plt.xlabel('Category ID')
plt.ylabel('views')
plt.xticks(rotation=90)
plt.show()



```

```python
#1. Count the number of videos in each category:

category_counts = df['category_id'].value_counts()
print(category_counts)
```

```python
#Calculate the average number of views per category:

average_views_per_category = df.groupby('category_id')['views'].mean()
print(average_views_per_category)
```

```python
#3. Identify the top 10 most viewed videos:

top_10_most_viewed = df.nlargest(10, 'views')
print(top_10_most_viewed)
```

```python
#4 Plot a histogram of the views distribution:

plt.hist(df['views'], bins=10)
plt.xlabel('Views')
plt.ylabel('Frequency')
plt.title('Distribution of Views')
plt.show()
```

```python
#5. Calculate the average number of likes per subscriber:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df["likes"] = pd.to_numeric(df['likes'], errors='coerce')
df["subscriber"] = pd.to_numeric(df['subscriber'], errors='coerce')
df['likes_per_subscriber'] = df['likes'] / df['subscriber']
average_likes_per_subscriber = df['likes_per_subscriber'].mean()
print(average_likes_per_subscriber)
```

```python
# 7. Identify the videos with the highest likes-to-dislikes ratio:

df["likes"] = pd.to_numeric(df['likes'], errors='coerce')
df["dislike"] = pd.to_numeric(df['dislike'], errors='coerce')
df['likes_dislikes_ratio'] = df['likes'] / df['dislike']
highest_likes_dislikes_ratio = df.nlargest(10, 'likes_dislikes_ratio')
print(highest_likes_dislikes_ratio)
```

```python
# 8. Plot a bar chart of the number of videos per channel:


plt.figure(figsize=(6, 6))
sns.countplot(data=df, x='channel_title')
plt.xticks(rotation=90)
plt.xlabel('Channel Title')
plt.ylabel('Number of Videos')
plt.title('Number of Videos per Channel')
plt.show()
```

```python
# 7. Calculate the average number of comments per day:

df["Trend_day_count"] = pd.to_numeric(df['Trend_day_count'], errors='coerce')
df["comment_count"] = pd.to_numeric(df['comment_count'], errors='coerce')
df['comments_per_day'] = df['comment_count'] / df['Trend_day_count']
average_comments_per_day = df['comments_per_day'].mean()
print(average_comments_per_day)
```

```python
#8. Plot a scatter plot of likes vs. dislikes:

plt.scatter(df['likes'], df['dislike'])
plt.xlabel('Likes')
plt.ylabel('Dislikes')
plt.title('Likes vs. Dislikes')
plt.show()
```

```python
 #9. Plot a line chart showing the trend of views over time:

df['Trend_day_count'] = pd.to_datetime(df['Trend_day_count'])
df = df.sort_values('Trend_day_count')
plt.plot(df['Trend_day_count'], df['views'])
plt.xlabel('Publish Date')
plt.ylabel('Views')
plt.title('Views Trend over Time')
plt.show()
```

```python
# 10. Plot a stacked bar chart showing the distribution of likes, dislikes, and comments across different categories:

category_counts = df['category_id'].value_counts()
category_likes = df.groupby('category_id')['likes'].sum()
category_dislikes = df.groupby('category_id')['dislike'].sum()
category_comments = df.groupby('category_id')['comment_count'].sum()

plt.bar(category_counts.index, category_likes, label='Likes')
plt.bar(category_counts.index, category_dislikes, bottom=category_likes, label='Dislikes')
plt.bar(category_counts.index, category_comments, bottom=category_likes+category_dislikes, label='Comments')

plt.xlabel('Category ID')
plt.ylabel('Count')
plt.title('Distribution of Likes, Dislikes, and Comments by Category')
plt.legend()
plt.show()
```
